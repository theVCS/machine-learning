```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
import pickle
import joblib
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import seaborn as sn
from sklearn import datasets
from sklearn.preprocessing import LabelEncoder
from sklearn import tree

df = pd.read_csv("salaries.csv")
df
```

```python
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
df.company = le_company.fit_transform(df.company)
df.job = le_job.fit_transform(df.job)
df.degree = le_degree.fit_transform(df.degree)
df
```

```python
model = tree.DecisionTreeClassifier()
model.fit(df[["company", "job", "degree"]], df["salary_more_then_100k"])
```

```python
model.score(df[["company", "job", "degree"]], df["salary_more_then_100k"])
```
